W3A3:
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input())
if n > 0 and (n & (n-1)) == 0:
    print('True')
else:
    print('False')
```

lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 


```python
a, b, c = map(int,input().split())
if b+c> a > abs(b-c) or a+b> c > abs(a-b) or a+c> b > abs(a-c):
 if a==b==c:
  print('Day la tam giac deu')
 elif a==b or b==c or a==c:
    print('Day la tam giac can')
 else:
  print('Day la tam giac thuong')
else:
  print('Day khong la tam giac')
```

W3A13:
[Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 

```python
a = int(input('Nhap so tien dien: '))
if 0 <= a <= 50: 
    s = a*1500 
if 51 <= a <= 100:
    s = (a-50)*2000 + 50*1500
if a >= 100:
    s = 50*1500 + 50*2000 + (a-100)*3000
print(s)


```
