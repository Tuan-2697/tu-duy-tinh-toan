W4A16: In các số chẵn chia hết cho 3 bé hơn 100

```python

for n in range(0, 101):
    if n % 2 == 0:
        if n % 3 == 0:
         print(n,end=" ")
```

W4A17: Nhập vào số nguyên dương a, in ra bảng cửu chương của a 

```python
a = int(input())
for i in range(1,10):
    print(f"{a} x {i} = {a*i}")
```

W4A18: Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b = map(int,input().split())

for i in range(1, min(a,b) +1):
    if a % i == 0 and b % i == 0:
        print(i,end=" ")
```

W4A20: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
n = int(input())
while n > 1:
    if n % 2 != 0:
        print('No')
        break
    n //= 2
else:
    print("YES")
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 


```python
n = int(input())
tong = 0
for i in range(n+1):
    tong += i 
print(tong)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = input()
chan = 0
le = 0
for i in range(0,len(n)):
    if int(n[i]) % 2 == 0:
        chan += 1
    else:
        le += 1
print(f"chan: {chan}, le:{le}")
```
