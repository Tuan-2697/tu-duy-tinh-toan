W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list


```python
n = list(map(int,input().split()))
gtln = n[0]
vtri = 0
for i in range(len(n)):
    if n[i] > gtln:
        gtln = n[i]
        vtri = i
print(gtln, vtri)
```

W6A20

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1

```python
n = list(map(int,input().split()))
x = int(input())
tim_duoc = False
for i in range(len(n)):
    if n[i] < x:
        print(f"so: {n[i]}, vtri: {i}")
        tim_duoc = True
if not tim_duoc:
    print('-1')
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
n = list(map(int,input().split()))
tang = True
giam = True
for i in range(len(n)-1):
    if n[i] <= n[i+1]:
        tang = False
    elif n[i] >= n[i+1]:
        giam = False
if tang:
     print('giam')
elif giam:
     print('tang')
else:
        print('vo_danh')
```
